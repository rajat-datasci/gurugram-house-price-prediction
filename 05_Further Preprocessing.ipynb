```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housing.csv')
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])
df.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_train_set.head()
```

```python
strat_test_set.head()
```

```python
df = strat_train_set.copy()
df1 = strat_test_set.copy()
```

```python
for dataset in (df, df1):
    dataset.drop('income_cat', axis=1, inplace=True)
```

```python
housing_features = df.drop('median_house_value', axis=1).copy()
housing_labels = df['median_house_value'].copy()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
housing_num = housing_features.select_dtypes(include=[np.number])

cleaned_data = imputer.fit_transform(housing_num)
```
